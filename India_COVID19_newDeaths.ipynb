```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```
df = pd.read_csv("/content/drive/My Drive/Academics/owid-covid-data.csv",
                 parse_dates=["date"])
df.head()
```

```
df = df[df.location == 'India']
df.head()
```

```
df.info()
```

```
fig, ax = plt.subplots()
ax.plot(df.date, df.new_cases);
```

```
df.sort_values(by=["date"], inplace=True)
df.head()
```

```
df_tmp = df.copy()
```

```
# Enriching our DataFrame by datetime attributes

df_tmp["Year"] = df_tmp.date.dt.year
df_tmp["Month"] = df_tmp.date.dt.month
df_tmp["Date"] = df_tmp.date.dt.day
df_tmp["DayOfWeek"] = df_tmp.date.dt.dayofweek
df_tmp["DayOfYear"] = df_tmp.date.dt.dayofyear
```

```
df_tmp.tail().T #, len(df_tmp.head().T)
```

```
df_tmp.info()
```

```
# iso_code, continent, location will be the same through out the data...we can drop them.
df_tmp.drop(["iso_code", "continent", "location"], axis=1, inplace=True)
```

```
# Find the columns which contains strings

for label, content in df_tmp.items():
    if pd.api.types.is_string_dtype(content):
        df_tmp[label] = content.astype("category").cat.as_ordered()

df_tmp.info()
```

# Filling missing values

```
df.isna().sum()
"new_cases_smoothed"
```

```
df_tmp.drop([
             "total_tests",
             "new_tests",
             "total_tests_per_thousand",
             "new_tests_per_thousand",
             "new_tests_smoothed",
             "new_tests_smoothed_per_thousand",
             "tests_per_case",
             "positive_rate",
             "tests_units"
            ],axis=1, inplace=True)
df_tmp.head()
```

```
labels = [
    "new_cases_smoothed",
    "total_deaths",
    "new_deaths",
    "new_deaths_smoothed",
    "total_cases_per_million",
    "new_cases_per_million",
    "new_cases_smoothed_per_million",
    "total_deaths_per_million",
    "new_deaths_per_million",
    "new_deaths_smoothed_per_million",
    "stringency_index"
]

for label in labels:
    df_tmp[label].fillna(df_tmp[label].mean(), inplace=True)
```

```
df_tmp.isna().sum()
```

```
df_tmp.dropna(subset=["total_cases", "new_cases"], inplace=True)
df_tmp.tail()
```

```
corr_matrix = df_tmp.corr()
fig, ax = plt.subplots(figsize=(20, 15))
ax = sns.heatmap(corr_matrix,
                 annot=True,
                 linewidths=0.5,
                 fmt=".2f",
                 cmap="RdYlGn");
```

```
from sklearn.model_selection import train_test_split
from datetime import datetime

X = df_tmp.drop(["new_cases", "date"], axis=1)
y = df_tmp.new_cases

x = df_tmp.drop(["new_cases"], axis=1)
x_train = x[x.date < np.datetime64("2020-07-21")]
x_test = x[x.date >= np.datetime64("2020-07-21")]


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
x_test
```

# Modelling

## 1. RandomForestRegressor

```
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(random_state=69)

model.fit(X_train, y_train)
```

```
score1 = model.score(X_test, y_test)
score1
```

```
y_preds = model.predict(X_test)
y_preds[:10]
```

```
from sklearn.metrics import mean_squared_log_error, mean_absolute_error, mean_squared_error


def rmse(y_true, y_preds):
    return np.sqrt(mean_squared_error(y_true, y_preds))

def rmsle(y_true, y_preds):
    return np.sqrt(mean_squared_log_error(y_true, y_preds))

def show_scores(model):
    test_preds = model.predict(X_test)
    
    scores = {
        "Test MAE": mean_absolute_error(y_test, test_preds),
        "Test RMSLE": rmsle(y_test, test_preds),
        "Test RMSE": rmse(y_test, test_preds),
    }
    return scores
```

```
rfr_scores = show_scores(model)
rfr_scores
```

```
# using GridSearchCV
from sklearn.model_selection import GridSearchCV

gs_grid = {
    "n_estimators": np.arange(10, 100, 10),
    "max_depth": [None, 3, 5, 10],
    "min_samples_split": np.arange(2, 10, 2),
    "min_samples_leaf": np.arange(1, 10 ,2),
    "max_features": [0.5, 1, "auto", "sqrt"]
}

gs_model = GridSearchCV(RandomForestRegressor(n_jobs=-1,
                                              random_state=69),
                        param_grid=gs_grid,
                        cv=5,
                        verbose=True)

gs_model.fit(X_train, y_train)
```

```
# using RandomSearchCV
from sklearn.model_selection import RandomizedSearchCV

rs_grid = {
    "n_estimators": np.arange(10, 1000, 10),
    "max_depth": [None, 3, 5, 10],
    "min_samples_split": np.arange(2, 10, 2),
    "min_samples_leaf": np.arange(1, 10 ,2),
    "max_features": [0.5, 1, "auto", "sqrt"]
}

rs_model = RandomizedSearchCV(RandomForestRegressor(n_jobs=-1,
                                              random_state=69),
                              param_distributions=rs_grid,
                              cv=5,
                              n_iter=69,
                              verbose=True)

rs_model.fit(X_train, y_train)
```

```
rs_model.best_params_
```

```
rs_score1 = rs_model.score(X_test, y_test)
rs_score1
```

## 2. Linear Regression

```
from sklearn.linear_model import LinearRegression

np.random.seed(69)

model2 = LinearRegression().fit(X_train, y_train)
```

```
score2 = model2.score(X_test, y_test)
score2
```

```
lr_scores = show_scores(model2)
lr_scores
```

## 3. Support Vector Regression (SVR)

- Return the coefficient of determination R^2 of the prediction.
- The best possible score is 1.0 and **it can be negative (because the model can be arbitrarily worse)**. A constant model that *always predicts the expected value of y*, disregarding the input features, would get a **R^2 score of 0.0**.

```
from sklearn.svm import SVR

np.random.seed(69)

model3 = SVR().fit(X_train, y_train)
```

```
score3 = model3.score(X_test, y_test)
score3
```

```
svr_scores = show_scores(model3)
svr_scores
```

## 4. AdaBoost Regressor

```
from sklearn.ensemble import AdaBoostRegressor

model4 = AdaBoostRegressor(random_state=69).fit(X_train, y_train)
```

```
score4 = model4.score(X_test, y_test)
score4
```

```
adaboost_scores = show_scores(model4)
adaboost_scores
```

## 5. XGBoost

```
from xgboost import XGBRegressor

np.random.seed(69)

model5 = XGBRegressor().fit(X_train, y_train)
score5 = model5.score(X_test, y_test)
score5
```

```
from sklearn.model_selection import GridSearchCV

grid = {
    "n_estimators": [50, 100, 150, 200],
    "learning_rate": [0.01, 0.1, 0.2, 0.3],
    "max_depth": np.arange(3, 10),
    "colsample_bytree": [i/10.0 for i in range(1, 3)],
    "gamma": [i/10.0 for i in range(3)],
    "k": [10]
}

gs_model5 = GridSearchCV(XGBRegressor(),
                         param_grid=grid,
                         cv=5,
                         verbose=True)
gs_model5.fit(X_train, y_train)
```

```
gs_model5.best_params_
```

```
gs_score5 = gs_model5.score(X_test, y_test)
gs_score5
```

```
xgboost_gs_scores = show_scores(gs_model5)
xgboost_gs_scores
```

```
model5.get_xgb_params()
```
